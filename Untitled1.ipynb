```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('metadata.txt',sep='\t')
data = data.set_index('File accession')
data.head()
```

```python
df = data[['Output type','Biosample term name','Biosample type','Biosample organism',\
           'Experiment target', 'Biological replicate(s)','Technical replicate']]
df = df.loc[df['Output type']=='unfiltered alignments']       
df = df.loc[df['Biosample organism']=='Homo sapiens']
df = df.loc[df['Biological replicate(s)']==1.0]
df = df.drop(['Biosample organism','Output type','Biological replicate(s)','Technical replicate','Biosample type'],axis=1)
df
```

```python
Experiment_Target = list(set(df['Experiment target'].values))
ExperimentTarget_Biosample = []
for i in range(len(df)):
    ExperimentTarget_Biosample.append(tuple(df.iloc[[i]].values[0]))
ExperimentTarget_Biosample = list(set(ExperimentTarget_Biosample))
#ExperimentTarget_Biosample
```

```python
ExperimentTarget_Biosample_Dict = {}
for i in Experiment_Target:
    ExperimentTarget_Biosample_Dict[i] = []
    for j in ExperimentTarget_Biosample:
        if j[1] == i:
            ExperimentTarget_Biosample_Dict[i].append(j[0])
#ExperimentTarget_Biosample_Dict
```

```python
ExperimentTarget_Biosample_Count_Dict = {}
for key in ExperimentTarget_Biosample_Dict.keys():
    ExperimentTarget_Biosample_Count_Dict[key] = len(ExperimentTarget_Biosample_Dict[key])
#ExperimentTarget_Biosample_Count_Dict
```

```python
# more than 5 biosample
min_num = 5
Target_dict = {}
for key in ExperimentTarget_Biosample_Count_Dict.keys():
    if ExperimentTarget_Biosample_Count_Dict[key] > min_num:
        Target_dict[key] = ExperimentTarget_Biosample_Dict[key]
#Target_dict
Target_list = list(Target_dict.keys())
#Target_list
len(Target_list)
```

```python
Target_count_dict = {}
for key in Target_dict.keys():
    Target_count_dict[key] = len(Target_dict[key])
#Target_count_dict
```

```python
Sorted_Target_list = []
Target_count_values = Target_count_dict.values()
Target_count_values = list(set(Target_count_values))
Target_count_values.sort(reverse=True)
for i in Target_count_values:
    for j in Target_list: 
        if Target_count_dict[j] == i:
            Sorted_Target_list.append(j)
#Sorted_Target_list
```

```python
Sorted_Target_list_ = []
for i in Sorted_Target_list:
    Sorted_Target_list_.append(i[:-6])
```

```python
result = pd.DataFrame(np.zeros([len(Sorted_Target_list),len(Sorted_Target_list)]))
result = result.set_index([pd.Index(Sorted_Target_list)])
result.columns = Sorted_Target_list
for i in Sorted_Target_list:
    for j in Sorted_Target_list:
        i_BioSample = ExperimentTarget_Biosample_Dict[i]
        j_BioSample = ExperimentTarget_Biosample_Dict[j]
        intersection = [k for k in i_BioSample if k in j_BioSample]
        result.loc[i][j] = len(intersection)
result = result.set_index([pd.Index(Sorted_Target_list_)])
result.columns = Sorted_Target_list_
result
```

```python
fig, ax = plt.subplots(figsize=(15,15)) 
sns.heatmap(result, annot=True,fmt=".3g",vmin=0, vmax=10,cmap="Blues",ax=ax)
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')
plt.yticks(rotation=0) 
plt.xticks(rotation=30) 
ax.set_ylim([23, 0])
```

```python
def common_sample(TF_list):
    common_sample_list = ExperimentTarget_Biosample_Dict[TF_list[0]]
    for i in TF_list:
        TF_BioSample = ExperimentTarget_Biosample_Dict[i]
        common_sample_list = list(set(TF_BioSample).intersection(set(common_sample_list)))
    return common_sample_list
```

```python
TF_list = Sorted_Target_list[:5]
common_sample(TF_list)
```
